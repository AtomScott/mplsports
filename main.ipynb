```python
import matplotlib.pyplot as plt
from mplsports import Pitch

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(20, 20))

# Flatten the 2D array of axes for easier indexing
axs = axs.flatten()

# Soccer
soccer_pitch = Pitch(sport="soccer", axis=True, tick=True)
soccer_pitch.draw(ax=axs[0])
axs[0].set_title("Soccer Pitch")

# Basketball
basketball_court = Pitch(sport="basketball", axis=True, tick=True)
basketball_court.draw(ax=axs[1])
axs[1].set_title("Basketball Court")

# Tennis
tennis_court = Pitch(sport="tennis", axis=True, tick=True)
tennis_court.draw(ax=axs[2])
axs[2].set_title("Tennis Court")

# Handball
handball_court = Pitch(sport="handball", axis=True, tick=True)
handball_court.draw(ax=axs[3])
axs[3].set_title("Handball Court")

# Adjust the layout and display the plot
plt.tight_layout()
plt.show()
```

```python
# example_custom_pitch.py
from mplsports.pitch.soccer import SoccerPitch
import matplotlib.pyplot as plt

# Create a soccer pitch with custom parameters
pitch = SoccerPitch(
    pitch_type='standard',
    half=False,
    pitch_color='#a8bc95',
    line_color='white',
    line_alpha=0.8,
    linewidth=3,
    linestyle='--',
    axis=True,
    tick=True,
    label=True,
    pad_left=5,
    pad_right=5,
    pad_bottom=2,
    pad_top=2
)

# Draw the pitch
ax = pitch.draw()

# Add a title
ax.set_title('Custom Soccer Pitch', fontsize=16)

# Show the plot
plt.show()
```

```python
# example_basketball_court.py
from mplsports.pitch.basketball import BasketballCourt
import matplotlib.pyplot as plt

# Create a basketball court instance
court = BasketballCourt(
    pitch_color='#f0f0f0',
    line_color='black',
    linewidth=2,
    half=False
)

# Draw the court
ax = court.draw()

# Add a title
ax.set_title('Basketball Court', fontsize=16)

# Show the plot
plt.show()
```

```python
# example_tennis_court.py
from mplsports.pitch.tennis import TennisCourt
import matplotlib.pyplot as plt

# Create a tennis court instance
court = TennisCourt(
    pitch_color='#008000',
    line_color='white',
    linewidth=2,
    line_zorder=1,
    axis=True
)

# Draw the court
ax = court.draw()

# Add a title
ax.set_title('Tennis Court', fontsize=16)

# Show the plot
plt.show()
```
